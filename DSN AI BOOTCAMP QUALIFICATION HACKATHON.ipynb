```python
train_set1 = train_set.copy()
test_set1 = test_set.copy()
```

```python
train_set1['form_field47'] = encoder.fit_transform(train_set['form_field47'])

test_set1['form_field47'] = encoder.fit_transform(test_set['form_field47'])
```

```python
train_set1.drop(['Applicant_ID', 'default_status'], axis=1, inplace=True)

test_set1.drop('Applicant_ID', axis=1, inplace=True)
```

```python
train_f = train_set1
train_label = train_set['default_status']

train_f.fillna(-99999, inplace=True)
test_set1.fillna(-99999, inplace=True)
```

```python

folds = 10
rand=42
skf = StratifiedKFold(n_splits=folds, shuffle=True, random_state= rand)
```

### ML Algorithms

####  Cross Validation For Loop With RandomForest

```python
pred_rf = []
sc_rf = []
roc_pred_rf = []
j = 1
for train_index, test_index in skf.split(train_f, train_label):
    x_train, x_test = train_f.loc[train_index], train_f.loc[test_index]
    y_train, y_test = train_label.loc[train_index], train_label.loc[test_index]
    
    

    
    
    
    rf_clf = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='entropy', max_depth=None, max_features='auto',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=400,
                       n_jobs=None, oob_score=False, random_state=42, verbose=100,
                       warm_start=False)

    rf_clf.fit(x_train, y_train)
    pred1 = rf_clf.predict_proba(x_test)[:, 1]
    pred_rf.append(pred1)
    print('roc_auc_score: ', roc_auc_score(y_test, pred1))
    roc_pred_rf.append(roc_auc_score(y_test, pred1))
    pred2 = rf_clf.predict_proba(test_set1)[:, 1]
    sc_rf.append(pred2)
    j += 1
```

```python
print('\nmean:', np.mean(roc_pred_rf, axis=0))
print(np.std(roc_pred_rf, axis=0))
```

```python
rf = np.mean(sc_rf, axis=0)
```

### XGBOOST

```python
pred_xgb = []
roc_pred_xgb = []
sc_xgb = []
j = 1
for train_index, test_index in skf.split(train_f, train_label):
    x_train, x_test = train_f.loc[train_index], train_f.loc[test_index]
    y_train, y_test = train_label.loc[train_index], train_label.loc[test_index]
    

    xgb_clf = xgboost.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=1.0, gamma=0.5,
              learning_rate=0.1, max_delta_step=0, max_depth=9,
              min_child_weight=10, missing=None, n_estimators=1000, n_jobs=1,
              nthread=1, objective='binary:logistic', random_state=0,
              reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
              silent=True, subsample=0.8, verbosity=1, eval_metric='auc')
   
    xgb_clf.fit(x_train, y_train, eval_set=[(x_train, y_train), (x_test, y_test)], early_stopping_rounds=300,  verbose=100)
    
    pred1 = xgb_clf.predict_proba(x_test)[:, 1]
    pred_xgb.append(pred1)
    print('roc_auc_score: ', roc_auc_score(y_test, pred1))
    roc_pred_xgb.append(roc_auc_score(y_test, pred1))
    pred2 = xgb_clf.predict_proba(test_set1)[:, 1]
    sc_xgb.append(pred2)
    j += 1        
```

```python
np.mean(roc_pred_xgb, axis=0)
```

```python
xgb = np.mean(sc_xgb, axis=0)
```

### LGB Classifier

```python
pred_lgb = []
roc_pred_lgb = []
sc_lgb = []
j = 1
for train_index, test_index in skf.split(train_f, train_label):
    x_train, x_test = train_f.loc[train_index], train_f.loc[test_index]
    y_train, y_test = train_label.loc[train_index], train_label.loc[test_index]
    
    lgb_clf = lgboost.LGBMClassifier(boosting_type='gbdt', class_weight=None, metric='auc',
               importance_type='split', learning_rate=0.01, max_bin=45,
               max_depth=14, min_child_weight=0.001, min_child_samples=36, 
                subsample_freq=1, subsample=0.739, colsample_bytree=0.8,
               min_split_gain=1.11, n_estimators=2800, n_jobs=-1, num_leaves=31, 
               objective='binary', random_state=42, reg_alpha=0.0, reg_lambda=0.0,
               silent=True, subsample_for_bin=200000)
    
    lgb_clf.fit(x_train, y_train, eval_set=[(x_test, y_test)], early_stopping_rounds=500, verbose=100, eval_metric='auc')
    
    pred1 = lgb_clf.predict_proba(x_test)[:, 1]
    pred_lgb.append(pred1)
    print('roc_auc_score: ', roc_auc_score(y_test, pred1))
    roc_pred_lgb.append(roc_auc_score(y_test, pred1))
    pred2 = lgb_clf.predict_proba(test_set1)[:, 1]
    sc_lgb.append(pred2)
    j += 1        
```

```python
np.mean(roc_pred_lgb, axis=0)
```

```python
lgb = np.mean(sc_lgb, axis=0)
```

### CatBoost Classifier

```python
pred_cat = []
roc_pred_cat = []
sc_cat = []
j = 1
for train_index, test_index in skf.split(train_f, train_label):
    x_train, x_test = train_f.loc[train_index], train_f.loc[test_index]
    y_train, y_test = train_label.loc[train_index], train_label.loc[test_index]
    
    
    cat_clf = catboost.CatBoostClassifier(n_estimators=2600, random_state=42, eval_metric='AUC',use_best_model=True,
                                    subsample = 0.739,  max_depth = 8, learning_rate = 0.011)
    
    
    cat_clf.fit(x_train, y_train, eval_set=[(x_train, y_train), (x_test, y_test)], early_stopping_rounds=500, verbose=100)
    
    pred1 = cat_clf.predict_proba(x_test)[:, 1]
    pred_cat.append(pred1)
    print('roc_auc_score: ', roc_auc_score(y_test, pred1))
    roc_pred_cat.append(roc_auc_score(y_test, pred1))
    pred2 = cat_clf.predict_proba(test_set1)[:, 1]
    sc_cat.append(pred2)
    j += 1        


```

```python
np.mean(roc_pred_cat, axis=0)
```

```python
cat = np.mean(sc_cat, axis=0)
```

### Blending The Models
